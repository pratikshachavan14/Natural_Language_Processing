```python
from transformers import pipeline
```

# Text Classification

```python
classifier = pipeline("text-classification")
```

```python
import pandas as pd
text = 'It was possible for him easily. So he did it.'
result = classifier(text)
df = pd.DataFrame(result)
df
```

```python
text = 'I know that it is not possible for you.'
result = classifier(text)
df = pd.DataFrame(result)
df
```

# Question Answering

```python
qa = pipeline("question-answering")
```

```python
text = 'Cash is a cute dog'
q = 'Who is cash?'
```

```python
a = qa(question=q, context=text)
a['answer']
```

```python
text = 'A is bigger than B'
q = 'Who is bigger?'
a = qa(question=q, context=text)
a['answer']
```

# Conversation

```python
from transformers import Conversation
converse = pipeline("conversational")
```

```python
conversation = Conversation("It was my first day today. Any suggestions?", max_length=50, num_return_sequences=3)
response = converse([conversation])
```

```python
response[1]['content']
```

# Visual question answering

```python
from PIL import Image
vqa_pipeline = pipeline("visual-question-answering")
```

```python
image = Image.open("dog.jpg")
question = "Who is there in image?"
vqa_pipeline(image, question, top_k=3)
```

```python
image = Image.open("dog.jpg")
question = "What is color of dog?"
vqa_pipeline(image, question, top_k=3)
```

```python
image = Image.open("rose-flower.jpg")
question = "What is there in image??"
vqa_pipeline(image, question, top_k=3)[0]['answer']
```

```python
image = Image.open("rose-flower.jpg")
question = "What is color of the flower??"
vqa_pipeline(image, question, top_k=3)[0]['answer']
```

# Text to speech

```python
pipe = pipeline('text-to-speech', model='suno/bark-small')
```

```python
text = "Hello, today i saw a beautiful flower in the garden."
output = pipe(text)
output
```
