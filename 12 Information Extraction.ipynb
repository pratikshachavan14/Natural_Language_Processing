```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
from nltk import FreqDist
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
url = urlopen("https://en.wikipedia.org/wiki/Harishchandragad")

data = url.read()   #read one time activity
data                #Ctrl + U to see source code
```

```python
bsoup = BeautifulSoup(data, 'html.parser')
bsoup
```

```python
bsoup.find('title')
```

```python
text = bsoup.get_text()
print(text)
```

```python
tokens = word_tokenize(text)
fdist = FreqDist(tokens)
fdist.plot(20)
```

```python
swords = stopwords.words('english')
tokens1 = [token for token in tokens if token.lower() not in swords]
tokens2 = [token.lower() for token in tokens1 if token.isalnum()]

fdist = FreqDist(tokens2)
fdist.plot(20)
```

```python
fdist.most_common(10)
```
