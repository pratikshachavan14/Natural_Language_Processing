```python
import numpy as np
from keras.utils import to_categorical
from keras.models import Sequential
from keras.layers import Dense, LSTM, SimpleRNN
```

```python
alphabets = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
```

```python
list(enumerate(alphabets))
```

```python
int_to_char = dict(enumerate(alphabets))
int_to_char
```

```python
char_to_int = dict((v, i) for i, v in enumerate(alphabets))
char_to_int
```

#### Data Preparation

```python
seq_length = 1
x = []
y = []
for i in range(len(alphabets) - seq_length):
  seq_in = [alphabets[i]]
  seq_out = alphabets[i + 1]
  x += [[char_to_int[seq_in[0]]]]
  y += [[char_to_int[seq_out]]]
  print(seq_in, '->', seq_out)
```

```python
x, y
```

#### Prepare the output variable

```python
y_new = to_categorical(y)     #converting to catergorical form i.e one hot encoding
y_new
```

#### Prepare input variables

```python
x = np.reshape(x, (25, 1, 1))
x
```

```python
x_new = x / 25   #normalizing the data
x_new
```

```python
x_new.shape, y_new.shape
```

#### Create the model

```python
model = Sequential()

model.add(Dense(25, input_shape = (1, 1), activation= 'relu'))
#model.add(LSTM(32))
model.add(SimpleRNN(32))
model.add(Dense(26, activation= 'softmax'))
```

#### Compile the model

```python
model.compile(loss= 'categorical_crossentropy', optimizer= 'adam', metrics= ['accuracy'])
```

#### Train the model

```python
model.fit(x_new, y_new, epochs = 1000, batch_size=1)
```

```python
model.evaluate(x_new, y_new, batch_size= 1)
```

#### Predict the character in sequence

```python
new = 'K'
new_int = char_to_int[new]
new_int = np.reshape(new_int, (1, 1, 1))
new_int = new_int / 25
new_int
```

```python
pred = model.predict(new_int)
pred = np.argmax(pred)
pred = int_to_char[pred]
pred
```

#### For predicting next 3 charcters


```python
seq_length = 3
x = []
y = []
for i in range(len(alphabets) - seq_length):
  seq_in = alphabets[i:i + seq_length]
  seq_out = alphabets[i + seq_length]
  x.append([char_to_int[char] for char in seq_in])
  y.append(char_to_int[seq_out])
  print(seq_in, '->', seq_out)
```

```python
x, y
```

```python
y_new = to_categorical(y)     #converting to catergorical form i.e one hot encoding
y_new
```

```python
x = np.reshape(x, (len(x), seq_length, 1))
x.shape
```

```python
x = x / 25   #normalizing the data
x
```

```python
model = Sequential()

model.add(Dense(25, input_shape = (3, 1), activation= 'relu'))
model.add(LSTM(128))
model.add(Dense(26, activation= 'softmax'))
```

```python
y_new = to_categorical(y)     #converting to catergorical form i.e one hot encoding
y_new.shape
```

```python
#Compile the model
model.compile(loss= 'categorical_crossentropy', optimizer= 'adam', metrics= ['accuracy'])
```

```python
#Train the model
model.fit(x, y_new, epochs = 1000, batch_size=1)
```

```python
model.evaluate(x, y_new, batch_size= 1)
```

#### Predict on new sequence

```python
new = 'IJK'
new_int = [char_to_int[char] for char in new]
new_int = np.reshape(new_int, (1, 3, 1))
new_int = new_int / 25
new_int
```

```python
pred = model.predict(new_int)
pred = np.argmax(pred)
pred = int_to_char[pred]
pred
```
