```python
from urllib.request import urlopen
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from bs4 import BeautifulSoup
from nltk import FreqDist
from nltk.tokenize import word_tokenize
```

```python
url = urlopen("https://en.wikipedia.org/wiki/Tiger")

data = url.read()   #read one time activity
data                #Ctrl + U to see source code
```

```python
bsoup = BeautifulSoup(data, 'html.parser')
bsoup
```

```python
bsoup.find('title')
```

```python
text = bsoup.get_text()
print(text)
```

```python
tokens = word_tokenize(text)
fdist = FreqDist(tokens)
fdist.plot(20)
```

```python
wnl = WordNetLemmatizer()
ps  = PorterStemmer()
swords = stopwords.words('english')
```

```python
#tokens1 = [ps.stem(token) for token in tokens if token.lower() not in swords]  #stemming 
tokens1 = [wnl.lemmatize(token, pos='n') for token in tokens if token.lower() not in swords] #lemmatizer
tokens2 = [token.lower() for token in tokens1 if token.isalnum()]
```

```python
fdist = FreqDist(tokens2)
fdist.plot(20)
```
